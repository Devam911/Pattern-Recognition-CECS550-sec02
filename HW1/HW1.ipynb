### 1. Describe what an identity matrix is. What does multiplying with an identity matrix do?

An identity matrix, often denoted as "I" is a square matrix in which all elements on the main diagonal are equal to 1, and all other elements are equal to 0.
When we multiply any matrix by an identity matrix the result is the original matrix itself. Identity matrices are often used because of this unique property as they act as a neutral element in matrix multiplication, preserving the original data structure when multiplied with other matrices.

### 1. We now want to retrieve the original image. Is matrix T invertible? How would you show that it is invertible?


To show that a matrix is invertible we calculate its determinant. If the determinant is zero then it is non-invertible matrix.

Formula for calculating the inverse of the matrix.
T^(-1) = (1 / det(T)) * adj(T)

So, it the determinant is zero the inverse is not defined.

```python
# 2. Compute the inverse and store the matrix in Tinv.

"""
Calculating the determinant of T using the det() function. Later checking the value of determinant and if its not zero then
calculate the inverse using the inv() function.
"""
det_T = np.linalg.det(T)

if det_T != 0:
    print("Matrix T is invertible.")
    Tinv = np.linalg.inv(T)
    
else:
    print("Matrix T is not invertible.")
```

```python
# 3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

"""
To perform Inverse operation we need to Multiply Tinv with transformed_image.
"""
retrieved_image = np.dot(Tinv, transformed_image)

# Plot the retrieved image
plt.figure(figsize=(5, 5))
plt.imshow(retrieved_image)
plt.title("Retrieved Original Image")
plt.show()
```

```python
"""
The matrix algebra only changes in creating the T_cols matrix and in the order of dot product. 
"""

# creating an array of size column
cols = np.arange(column)

# Shuffling the indices of the array
np.random.shuffle(cols)

# crearting a T matrix which is the shuffled version of the identity matrix by columnns array.
T_cols = identity_matrix[:,cols]

# Printing the shape of the T_cols matrix
print("T_cols Shape",T_cols.shape)

# Created a transformed image by doing dot product of image and the T matrix.
transformed_image_cols = np.dot(img_average_channel,T_cols)

# Plotting the scrambled image
plt.figure(figsize=(5, 5))
plt.imshow(transformed_image_cols)
plt.title("Transformed Image by columns")
plt.show()
```

```python
# Calculating the determinant and later calculating the inverse of the matrix 
det_T_cols = np.linalg.det(T_cols)
if det_T_cols != 0:
    T_cols_inv = np.linalg.inv(T_cols)
    print("Matrix T_cols is invertible.")
else:
    print("Matrix T_cols is not invertible.")

# Calculating the dot product of transformed image and inverse of T to retrive the image.
retrieved_image_cols = np.dot(transformed_image_cols, T_cols_inv)

# Plotting the retrived image
plt.figure(figsize=(5, 5))
plt.imshow(retrieved_image_cols)
plt.title("Retrieved Original Image")
plt.show()
```
